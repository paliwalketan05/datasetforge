```python
import pandas as pd
```

```python
import requests
from config import API_KEY

url = "https://imdb236.p.rapidapi.com/api/imdb/top250-movies"

headers = {
	"x-rapidapi-key": API_KEY,
	"x-rapidapi-host": "imdb236.p.rapidapi.com",
	"Content-Type": "application/json"
}

response = requests.get(url, headers=headers)

print(response.json())
```

```python
df=pd.DataFrame(response.json())[['primaryTitle','originalTitle','type','description','contentRating','interests']]
```

```python
df.isnull().sum()
```

```python
df.fillna("NaN")
```

```python
df.columns = [
    "movie_title",
    "original_title",
    "type",
    "description",
    "content_rating",
    "genres"
]
```
